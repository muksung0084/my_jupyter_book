#### q cut 범위별 range 자르기 

```python
# q cut 범위별 range 자르기 
sample['range'] = pd.qcut(sample['a'],q=5)

# 소수점 없애기 precision 
sample['range_2'] = pd.qcut(sample['a'],q=5,precision=0)

# range를 범주화 시키기 
sample['range_labels'] = pd.qcut(sample['a'],q=5,labels=range(5))

# duplicates drop 추가
sample['range_labels_2'] = pd.qcut(sample['a'],q=5,duplicates='drop')
```
